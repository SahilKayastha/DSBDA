### Import Required Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Read Dataset

```python
df = sns.load_dataset('titanic')			# Load dataset
df.head()
```

### **1) Histogram:**

`sns.histplot()`

* **Type of Plot**: Histogram (with optional KDE)
* **Purpose**: Visualizes the distribution of a single continuous variable by dividing it into bins.
* **Use Case**: Analyzing the frequency distribution of numerical data, such as passenger fares.

**Key Parameters**:

| Parameter  | Description                                                                                 |                           |
| ---------- | ------------------------------------------------------------------------------------------- | ------------------------- |
| `data`     | Dataset for plotting.                                                                       |                           |
| `x` or `y` | Variable to plot on the respective axis.                                                    |                           |
| `bins`     | Number of bins for the histogram.                                                           |                           |
| `kde`      | If True, overlays a Kernel Density Estimate.                                                |                           |
| `hue`      | Categorical variable for color encoding.                                                    |                        

```python
sns.histplot(data=df, x='fare', bins=10, kde=True)
```

**Observation:** Most passengers paid low fares, with a right-skewed distribution indicating fewer high-paying passengers.

```python
sns.histplot(data=df, x='age', bins=10, kde=True)
```

**Observation:** Passenger ages are concentrated between 20 and 40 years, with fewer children and elderly individuals.

### **2) JointPlot:**

`sns.jointplot()`

* **Type of Plot**: Joint Plot (combines scatter plot with marginal histograms)
* **Purpose**: Displays the relationship between two variables along with their individual distributions.
* **Use Case**: Analyzing bivariate relationships, such as age versus fare.

**Key Parameters**:

| Parameter | Description                                                     |   |
| --------- | --------------------------------------------------------------- | - |
| `data`    | Dataset for plotting.                                           |   |
| `x`, `y`  | Variables to plot on the x and y axes.                          |   |
| `kind`    | Type of plot: `'scatter'`, `'kde'`, `'hist'`, `'hex'`, `'reg'`. |   |
| `hue`     | Categorical variable for color encoding.                        |   |
| `height`  | Size of the figure (in inches).                                 |   |
| `ratio`   | Ratio of joint axes height to marginal axes height.             |   |


```python
sns.jointplot(data=df, x='age', y='fare', kind='scatter')
```

**Observation:** Younger and middle-aged passengers mostly paid lower fares, with a few older passengers paying premium prices.

### **3) BarPlot:**

`sns.barplot()`

* **Type of Plot**: Bar Plot
* **Purpose**: Shows the central tendency (e.g., mean) of a numerical variable for each category.
* **Use Case**: Comparing average values across categories, such as average age by sex.

**Key Parameters**:

| Parameter   | Description                                                   |   |
| ----------- | ------------------------------------------------------------- | - |
| `data`      | Dataset for plotting.                                         |   |
| `x`, `y`    | Variables for the x and y axes.                               |   |
| `hue`       | Categorical variable for color encoding.                      |   |
| `estimator` | Function to estimate central tendency (e.g., `np.mean`).      |   |
| `ci`        | Size of confidence intervals to draw around estimated values. |   |
| `capsize`   | Width of the caps on error bars.                              |   |


```python
sns.barplot(data=df, x='sex', y='age')
plt.title("Average Age by Sex")
```

**Observation:** On average, male passengers were slightly older than female passengers.

### **4) countPlot:**

`sns.countplot()`

* **Type of Plot**: Count Plot
* **Purpose**: Displays the count of observations in each categorical bin.
* **Use Case**: Visualizing the number of occurrences for each category, such as survival counts by sex.

**Key Parameters**:

| Parameter  | Description                              |                           
| ---------- | ---------------------------------------- |
| `data`     | Dataset for plotting.                    |                           
| `x` or `y` | Categorical variable to plot.            |                           
| `hue`      | Categorical variable for color encoding. |                          
| `order`    | Order to plot the categorical levels.    |                           
| `palette`  | Color palette for the plot.              | 


```python
sns.countplot(data=df, x='survived', hue='sex')
plt.title("Surival rate")
```

**Observation:** Females had a higher survival rate compared to males, indicating gender-based survival differences.

### **5) BoxPlot:**

`sns.boxplot()`

* **Type of Plot**: Box Plot
* **Purpose**: Summarizes the distribution of a dataset using five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.
* **Use Case**: Identifying outliers and comparing distributions across categories, such as age distribution by sex.([GeeksforGeeks][5])

**Key Parameters**:

| Parameter | Description                              |   |
| --------- | ---------------------------------------- | - |
| `data`    | Dataset for plotting.                    |   |
| `x`, `y`  | Variables for the x and y axes.          |   |
| `hue`     | Categorical variable for color encoding. |   |
| `order`   | Order to plot the categorical levels.    |   |
| `palette` | Color palette for the plot.              |   |


```python
sns.boxplot(data=df, x='sex', y='age')
```

**Observation:** Males show a wider age range and higher median age compared to females.

```python
sns.boxplot(x='sex', y='age', data=df, hue='survived')
```

**Observation:** Among survivors, females were generally younger, while males showed a broader age range in both survival outcomes.

### **6) StripPlot:**

`sns.stripplot()`

* **Type of Plot**: Strip Plot
* **Purpose**: Draws a scatter plot where one variable is categorical.
* **Use Case**: Displaying all individual observations, especially useful for small datasets.

**Key Parameters**:

| Parameter | Description                                                                               |                                               |
| --------- | ----------------------------------------------------------------------------------------- | --------------------------------------------- |
| `data`    | Dataset for plotting.                                                                     |                                               |
| `x`, `y`  | Variables for the x and y axes.                                                           |                                               |
| `hue`     | Categorical variable for color encoding.                                                  |                                               |
| `jitter`  | Adds random noise to the position of points along the categorical axis to reduce overlap. |                                               |
| `dodge`   | When True, separates the strips for different hue levels.                                 |                                               |
| `palette` | Color palette for the plot.                                                               | 



```python
sns.stripplot(data=df, x='sex', y='age', jitter=True)
```

**Observation:** Age distribution across sexes shows a dense clustering of young adults, with a slightly higher spread among males.

### Observe the distribution of fares and any insights related to passenger demographics or class.

```python
sns.boxplot(data=df, x='embarked', y='fare')
plt.title('Fare Distribution by Embarkation Port')
plt.xlabel('Embarkation Port')
plt.ylabel('Fare')
plt.show()
```

**Observation:**
- Passengers who embarked from port 'C' (Cherbourg) tend to have paid higher fares, with a wider range and more outliers.
- Ports 'Q' (Queenstown) and 'S' (Southampton) show lower median fares with fewer high-fare outliers.

```python
sns.boxplot(data=df, x='sex', y='fare')
plt.title('Fare Distribution by Sex')
plt.xlabel('Sex')
plt.ylabel('Fare')
plt.show()
```

**Observation:**
- Both males and females show similar fare distributions, but females have a slightly higher median fare.
- There are notable outliers in both groups, indicating that some individuals paid significantly higher fares.
