* **Compare distributions**: Analyze feature distributions and identify any visible outliers or skewness.

### Import Required Libraries

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

### Read Dataset

```python
df = sns.load_dataset('iris')
df.head()
```

### Inspect Dataset Structure

```python
print("Column Names and Data Types:\n", df.dtypes)
print("\nStatistical Summary:\n", df.describe())
```

### List Features and Classify as Numeric or Nominal

```python
# List all columns
features = df.columns.tolist()
print("All Features:", features)
```

```python
# Separate numeric and nominal features
numeric_features = df.select_dtypes(include='number').columns.tolist()
nominal_features = df.select_dtypes(exclude='number').columns.tolist()
```

```python
print("\nNumeric Features:", numeric_features)
print("Nominal Features:", nominal_features)
```

### Plot Histograms for All Numeric Features

```python
df[numeric_features].hist(bins=15, figsize=(10, 6), layout=(2, 2), edgecolor='black')
plt.suptitle('Histograms of Numeric Features', fontsize=16)
plt.tight_layout()
plt.show()
```

### Plot Box Plots for Each Numeric Feature

```python
# Set plot size
plt.figure(figsize=(12, 6))

# Create boxplots
for i, feature in enumerate(numeric_features, 1):
    plt.subplot(1, 4, i)
    sns.boxplot(y=df[feature])
    plt.title(feature)

plt.suptitle('Box Plots of Numeric Features', fontsize=16)
plt.tight_layout()
plt.show()
```

### Observations:

1. **Histograms**:
   - Most numeric features show a roughly normal distribution.
   - 'petal_length' and 'petal_width' show skewed distributions with clear species-based separation.

2. **Box Plots**:
   - 'sepal_width' contains visible outliers.
   - 'petal_length' and 'petal_width' have greater spread and more pronounced variation.
   - 'sepal_length' shows a fairly symmetrical distribution.

3. **Skewness and Outliers**:
   - 'sepal_width' is slightly left-skewed and contains outliers.
   - 'petal_width' has low values with a long tail to the right.
