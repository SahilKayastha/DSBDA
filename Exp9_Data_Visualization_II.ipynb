### Import Required Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### Read Dataset

```python
df = sns.load_dataset('titanic')			# Load dataset
df.head()
```

### Understand Data

```python
df.info()
```

```python
df.describe()
```

### Plot the Box Plot

`sns.boxplot()`

* **Type of Plot**: Box Plot
* **Purpose**: Summarizes the distribution of a dataset using five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum.
* **Use Case**: Identifying outliers and comparing distributions across categories, such as age distribution by sex.([GeeksforGeeks][5])

**Key Parameters**:

| Parameter | Description                              |   |
| --------- | ---------------------------------------- | - |
| `data`    | Dataset for plotting.                    |   |
| `x`, `y`  | Variables for the x and y axes.          |   |
| `hue`     | Categorical variable for color encoding. |   |
| `order`   | Order to plot the categorical levels.    |   |
| `palette` | Color palette for the plot.              |   |


```python
sns.boxplot(data= df, x='sex', y='age', hue='survived')

# Add title and axis labels
plt.title('Age Distribution by Sex and Survival Status on Titanic', fontsize=14)
plt.xlabel('Sex', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.legend(title='Survived', labels=['No (0)', 'Yes (1)'])

#Display the plot
plt.show()
```

### Observations
Inferences from the box plot:
1. **Female Passengers**:
   - Those who survived (Survived = 1) had a slightly wider age range and included more younger individuals.
   - Survivors among females generally had lower age variance compared to non-survivors.

2. **Male Passengers**:
   - A larger number of young males did not survive.
   - The median age of surviving males was lower than that of non-surviving males, indicating younger males had slightly better chances.

3. **General Trend**:
   - Children (ages under 10) were more likely to survive across both sexes.
   - Survival among older age groups (especially males) was relatively rare.

Conclusion: Age played a notable role in survival chances, particularly favoring younger passengers and females.
